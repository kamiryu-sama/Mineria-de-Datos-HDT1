```

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('tmdb-movies.csv')
dataset.columns
```

# **Películas que ganaron más?**

```
f = dataset.groupby(['original_title']).mean().sort_values(['revenue'], ascending=False)
f.head(5)
```

# **Películas que recuperaron presupuesto?**

```
dataset['Profits'] = dataset['revenue'] - dataset['budget']
profits = dataset.groupby(['original_title']).mean().sort_values(['Profits'], ascending=False)
profits.head(5)
```

# **Relación entre calificación y cuánto ganaron?**

```
dataset['able'] = dataset['vote_count'] > 2000;
p = dataset.groupby(['original_title']).mean().sort_values(['able', 'vote_average'], ascending=False)
p.head(5)
```

# **Géneros que generaron más dinero?**

```
p = dataset.groupby(['genres']).mean().sort_values(['Profits'], ascending=False)
p.head(5)
```

# **Director más exitoso (popularidad, ingresos)?**

```
p = dataset.groupby(['original_title']).mean().sort_values(['vote_count', 'Profits'], ascending=False)
p.head(5)
```

# **Tiene la duración que ver con la taquilla?**

```
p = dataset.groupby(['original_title']).mean().sort_values(['runtime', 'revenue'], ascending=False)
p.head(5)
```

# **Actores que tienen más taquilla?**

```
f = dataset.groupby(['cast']).mean().sort_values(['revenue'], ascending=False)
f.head(5)
```

# **Año más propicioso para películas?**

```
a = dataset.groupby(['release_year']).mean().sort_values(['popularity'], ascending=False)
a.head(5)
```

# **Relación entre género y calificación?**

```
dataset['able'] = dataset['vote_count'] > 2000;
p = dataset.groupby(['original_title']).mean().sort_values(['revenue', 'vote_average'], ascending=False)
p.head(5)
```


# **Géneros que predominan cada año?**

```
genre = dataset.groupby(['genres','release_year']).size().reset_index().groupby('genres')[[0]].max()
genre.head(5)
```

# **Género que tiene más pérdidas?**

```
p = dataset.groupby(['genres']).mean().sort_values(['Profits'],ascending=True)
p.head(5)
```
